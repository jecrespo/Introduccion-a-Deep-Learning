```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loss_crossentropy(y_true, y_pred):
    return -(y_true*np.log(y_pred) + (1-y_true)*np.log(1-y_pred))
```

Jhinge​=i∑​max(0,1−yi​∗hθ​(xi​))

```python
def loss_hinge(y_true, y_pred):
    l = []
    for yp in y_pred:
        l.append(max(0, 1-y_true*yp))
    return l
```

```python
def loss_hinge_squared(y_true, y_pred):
    l = []
    for yp in y_pred:
        l.append(max(0, 1-y_true*yp) ** 2)
    return l
```

```python
y_true = 1
y_pred = np.linspace(-1,1,50)
plt.figure()
plt.plot(y_pred, loss_crossentropy(y_true, y_pred), 'k', label='crossentropy')
plt.plot(y_pred, loss_hinge(y_true, y_pred), 'r', label='hinge')
plt.plot(y_pred, loss_hinge_squared(y_true, y_pred), 'b', label='hinge_squared')
plt.ylabel('$J$')
plt.xlabel('$y_{pred}$')
plt.title(f'$y_{{true}}$: {y_true}')
plt.legend()
```
