```python
import tensorflow as tf
tf.random.set_seed(0) #pongo una semilla de aleatorio para obtener todos el mismo resultado, es el punto de partida de los pesos
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train, y_train),(x_test, y_test) = mnist.load_data()
```

```python
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
#defino el modelo de red neuronal o arquitectura
#cambio relu por swish
model = tf.keras.models.Sequential([ 
    tf.keras.layers.Flatten(input_shape=(28, 28)), #primera capa aplasta las imagenes, las convierte en un vector y le digo el tamaño 28x28 pixeles, es la entrada que son 784 entradas
    tf.keras.layers.Dense(512, activation=tf.nn.swish), #segunda capa densa de 512 neuronas con función de activació relu
    tf.keras.layers.Dropout(0.2), #capa de dropout del 20%, en cada iteración apaga el 20% de las neuronas al azar para que el resto aprendan de forma robusta
    tf.keras.layers.Dense(10, activation=tf.nn.softmax) #capa densa de 10 neurona para las 10 salidas posibles que son los número del 0 al 9
])
```

```python
x_train.shape
```

```python
#compilo el modelo donde le añado: algoritmo de optimización (descenso de gradiente),función de perdida (entropia cruzada) y métrica para guiar el proceso de optimización (precisión)
model.compile(optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=5,
         validation_data=(x_test,y_test),
         validation_steps = len(x_test)) #lo ejecuto 5 veces buscando los pesos, accuracy es el % de aciertos pero son los de entrenamiento que no dice nada, lo importante son los de test
                                        #Y le paso el conjunto de los datos de validación
```

```python
model.evaluate(x_test, y_test) #lo testeo
```
