```python
import numpy as np
n_inputs = 784
n_layers = 100
x = np.random.randn(n_inputs)
for layer in range(n_layers):
    weights = np.random.randn(n_inputs, n_inputs)
    x = weights @ x
print(x.mean(), x.std())
```
